```python
import geopandas as gpd
import matplotlib.pyplot as plt
import os
import pandas as pd

# Set the path to the directory containing your shapefiles
directory = "src/"  # Replace with your directory path

# List of shapefiles you want to load
shapefiles = [
    "AC_1.shp", "AC_2.shp", "PCA_CANT.shp", "VETOS_TODOS.shp", "ZC_a.shp", "ZC.shp",
    "ZCOR_1.shp", "ZCOR_2.shp", "ZCOR_3.shp", "ZCOR_a.shp", "ZC_ZEIS.shp", "ZDE_1.shp",
    "ZDE_2.shp", "ZEIS_1.shp", "ZEIS_2.shp", "ZEIS_3.shp", "ZEIS_4.shp", "ZEIS_5.shp",
    "ZEMP.shp", "ZEM.shp", "ZEPAM.shp", "ZEP.shp", "ZER_1.shp", "ZER_2.shp", "ZER_a.shp",
    "ZEU_a.shp", "ZEUP_a.shp", "ZEUP_u.shp", "ZEU_u.shp", "ZM_a.shp", "ZMIS_a.shp",
    "ZMIS_u.shp", "ZM_u.shp", "ZOE.shp", "ZPDS_r.shp", "ZPDS_u.shp", "ZPI_1.shp",
    "ZPI_2.shp", "ZPR.shp"
]

# Initialize an empty list to hold GeoDataFrames
gdfs = []

# Loop through the shapefiles, read them, and append to the list
for shp in shapefiles:
    filepath = os.path.join(directory, shp)
    try:
        # Read the shapefile into a GeoDataFrame
        gdf = gpd.read_file(filepath)
        gdfs.append(gdf)
    except Exception as e:
        print(f"Could not read {shp}: {e}")

# Concatenate all GeoDataFrames into one
if gdfs:
    combined_gdf = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True))

    # Plot the combined GeoDataFrame
    combined_gdf.plot(figsize=(15, 10))
    plt.show()
else:
    print("No valid shapefiles found.")
```
